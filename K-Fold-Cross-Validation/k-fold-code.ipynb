```python
from sklearn.datasets import load_digits
digit_dataset=load_digits()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digit_dataset.data,digit_dataset.target,test_size=0.3)
```

```python
from sklearn.model_selection import StratifiedKFold
folds=StratifiedKFold(n_splits=3)

def get_score(model, x_train,x_test,y_train,y_test):
    model.fit(x_train, y_train)
    return model.score(x_test, y_test)

score_logistic_reg=[]
score_svm=[]
score_random_forest=[]
for train_index,test_index in folds.split(digit_dataset.data,digit_dataset.target):
    x_train,x_test,y_train,y_test=digit_dataset.data[train_index],digit_dataset.data[test_index],digit_dataset.target[train_index],digit_dataset.target[test_index]
    score_logistic_reg.append(get_score(LogisticRegression(max_iter=2000),x_train,x_test,y_train,y_test)*100)
    score_svm.append(get_score(SVC(gamma='auto'),x_train,x_test,y_train,y_test)*100)
    score_random_forest.append(get_score(RandomForestClassifier(n_estimators=40),x_train,x_test,y_train,y_test)*100)
```

cross_val_score function
------------------------

Logistic regression model performance using cross_val_score

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(solver='saga',max_iter=2000),digit_dataset.data,digit_dataset.target,cv=3)
```

SVM model performance using cross_val_score

```python
cross_val_score(SVC(gamma='auto'),digit_dataset.data,digit_dataset.target,cv=3)
```

Random Forest model performance using cross_val_score

```python
cross_val_score(RandomForestClassifier(n_estimators=40),digit_dataset.data,digit_dataset.target,cv=3)
```
