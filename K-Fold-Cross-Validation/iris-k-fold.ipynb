```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris_dataset=load_iris()
```

```python
iris_dataset.data
```

```python
iris_dataset.target
```

```python
from sklearn.model_selection import StratifiedKFold
folds=StratifiedKFold(n_splits=3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris_dataset.data,iris_dataset.target,test_size=0.3)
```

```python
def get_score(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    return model.score(x_test,y_test)
```

```python
score_lr=[]
score_svc=[]
score_rf=[]
for train_index,test_index in folds.split(iris_dataset.data,iris_dataset.target):
    x_train,x_test,y_train,y_test=iris_dataset.data[train_index],iris_dataset.data[test_index],iris_dataset.target[train_index],iris_dataset.target[test_index]
    score_lr.append(get_score(LogisticRegression(max_iter=2000),x_train,x_test,y_train,y_test))
    score_svc.append(get_score(SVC(gamma='auto'),x_train,x_test,y_train,y_test))
    score_rf.append(get_score(RandomForestClassifier(n_estimators=40),x_train,x_test,y_train,y_test))
```

```python
score_lr
```

```python
score_svc
```

```python
score_rf
```

cross_val_score function
------------------------

Logistic Regression

```python
from sklearn.model_selection import cross_val_score
lr_score=cross_val_score(LogisticRegression(max_iter=2000),iris_dataset.data,iris_dataset.target,cv=3)
lr_score
```

```python
np.average(lr_score)
```

Support Vector Machine (SVM)

```python
svc_score=cross_val_score(SVC(gamma='auto'),iris_dataset.data,iris_dataset.target,cv=3)
svc_score
```

```python
np.average(svc_score)
```

Random Forest

```python
rf_score=cross_val_score(RandomForestClassifier(n_estimators=10),iris_dataset.data,iris_dataset.target,cv=3)
rf_score
```

```python
np.average(rf_score)
```

Decision Tree

```python
decision_scores = cross_val_score(DecisionTreeClassifier(), iris_dataset.data, iris_dataset.target,cv=3)
decision_scores
```
