Random Forest = Bagging + Feature Randomness

Random Forest is the most popular algorithm that uses bagging internally.

Random Forest is a supervised machine learning algorithm that uses an ensemble of decision trees to make more accurate and stable predictions.

It is based on two main ideas:

    1)Bagging (Bootstrap Aggregation)

    2)Random feature selection at each split

This makes Random Forest powerful, less overfitted, and more robust than a single decision tree.

üå≥ Simple Explanation
------------------------
Think of a Random Forest as:

    ‚ÄúA group of decision trees. Each tree votes, and the forest decides the final answer.‚Äù


üîç How Random Forest Works
1. Bootstrapping the Data (Bagging)
--------------------------------------

    From the original dataset (N rows), the algorithm creates multiple new datasets (each of size N) by sampling with replacement.

    Each dataset is used to train one tree.

2. Random Feature Selection
----------------------------------

    When a tree tries to split a node, it does NOT look at all features.

    Instead, it picks a random subset of features and chooses the best split among those.(for eg: the algorithm selects a random subset of features at each split.Then it finds the best feature among them to become the root node (or next node).)

    This randomness creates diverse trees, which improves accuracy and prevents overfitting.

3. Aggregation
---------------

    Classification: majority vote of all trees

    Regression: average of all trees' predictions


üß† Intuition Behind Random Forest
----------------------------------------
A single tree is unstable.
Small changes in data ‚Üí big changes in outputs.(Low Bias,High variance)

A forest averages many diverse trees ‚Üí stable + generalizable predictions.


üì¶ Example

Suppose you want to predict whether a customer will buy a product.

Random Forest will:

    Create, say, 100 bootstrapped datasets

    Train 100 decision trees

    Combine predictions:

        -If 70 trees say Yes, 30 say No ‚Üí final prediction = Yes


üìä Random Forest Hyperparameters (Important)
--------------------------------------------

    n_estimators: number of trees

    max_depth: maximum depth of each tree

    max_features: number of random features per split

    min_samples_split: minimum samples to split a node
    (min_samples_split refers to the number of rows (data points) required in a node before a decision tree is allowed to split that node into branches.)

    bootstrap: whether to use bootstrapping (usually True)
