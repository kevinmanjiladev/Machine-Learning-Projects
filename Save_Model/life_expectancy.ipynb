```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
data={'Actual_Value':y_test,"Predicted_Value":y_pred}
df1=pd.DataFrame(data)
df1['difference']=y_test-y_pred
df1
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
mea=mean_absolute_error(y_test,y_pred)
mea
```

```python
mape=mean_absolute_percentage_error(y_test,y_pred)
mape
```

0.04728135210540849 is a good mean absolute percentage error

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
rmse=root_mean_squared_error(y_test,y_pred)
rmse
```

```python
r2=r2_score(y_test,y_pred)
r2*100
```

82.62618750622222 is a good r2_Score

What is Pickle?

Pickle is a Python module that converts a Python object into a byte stream and stores it in a file, and later reconstructs the object back into memory.
