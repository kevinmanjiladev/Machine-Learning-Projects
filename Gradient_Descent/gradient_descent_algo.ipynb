```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df=pd.read_csv('test_scores.csv')
x=np.array(df['math'])
y=np.array(df['cs'])
```

```python
def sklearn_gradient():
    from sklearn.linear_model import LinearRegression
    model=LinearRegression()
    model.fit(df[['math']],df['cs'])
    return model.coef_,model.intercept_
```

```python
def gradient_descent_algo(x,y):
    m_current=0
    b_current=0
    iterations=1000000
    n=len(x)
    cost_previous=0
    learning_rate=0.0002
    for i in range(iterations):
        y_pred=m_current*x+b_current
        cost= (1/n)*sum([val**2 for val in (y-y_pred)])
        md=-(2/n)*sum(x*(y-y_pred))
        bd=-(2/n)*sum(y-y_pred)
        m_current=m_current-learning_rate*md
        b_current=b_current-learning_rate*bd
        if math.isclose(cost,cost_previous,rel_tol=1e-20):
            break
        cost_previous=cost
    # return f'm:{[m_current]},b:{[b_current]},cost:{[cost]},iteration:{[i]}'
    print(cost_previous)
    return m_current,b_current
    
```

```python
m_sklearn,b_sklearn=sklearn_gradient()
print("Using sklearn: Coef {} Intercept {}".format(m_sklearn,b_sklearn))
```

```python
m,b=gradient_descent_algo(x,y)
print("Using gradient descent function: Coef {} Intercept {}".format(m, b))
```
