Hyperparameter tuning
----------------------
Hyperparameter tuning is the process of finding the best set of hyperparameters for a machine-learning model so it performs as well as possible.

üîß What are hyperparameters?

Hyperparameters are settings you choose before training the model.
They control how the model learns, but they are not learned from data.

Examples:

    K in K-Means

    Learning rate in neural networks

    Number of trees in Random Forest

    Max depth in Decision Trees

    C and gamma in SVM

    k (neighbors) in KNN

    Alpha in Lasso/Ridge regression

These values affect performance, so choosing them properly is important.

üéØ What is hyperparameter tuning?
---------------------------------
It is the process of testing many combinations of hyperparameters
to find the combination that gives the best accuracy / lowest error.
