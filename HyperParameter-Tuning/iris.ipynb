The RBF (Radial Basis Function) kernel creates non-linear decision boundaries.

Because:

It handles non-linear data extremely well.

It can create complex shapes for classification.

GridSearchCV
---

GridSearchCV is a tool in scikit-learn used to find the best hyperparameters for a machine-learning model by trying every possible combination from a grid (a list) of values you provide.

It stands for:

Grid Search + Cross Validation

üß† Why we need GridSearchCV?
---
Models depend on hyperparameters like:

k in KNN

C, gamma in SVM

max_depth in Decision Trees

n_estimators in Random Forest

Choosing them manually is difficult.
GridSearchCV tests many combinations and picks the one with best accuracy.

‚öôÔ∏è How GridSearchCV Works (Simple Steps)
---
Step 1: You define the model

Example: SVM

Step 2: You define the parameter grid

Example:

    param_grid = {
        'C': [0.1, 1, 10],
        'gamma': [0.001, 0.01, 0.1],
        'kernel': ['rbf']
    }
Step 3: GridSearchCV tries ALL combinations
Like:

C=0.1, gamma=0.001

C=0.1, gamma=0.01

C=0.1, gamma=0.1

C=1, gamma=0.001

‚Ä¶ etc.

Step 4: For each combination, it performs cross-validation
This prevents overfitting.

Step 5: Returns the best parameters and best model

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)
clf.fit(iris_dataset.data,iris_dataset.target)
clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
# df
df[['param_C','param_kernel','mean_test_score','rank_test_score']]
```

```python
clf.best_score_
```

```python
clf.best_params_
```

Disadvantage of GridSearchCV:
---
Computational cost of GridSearchCV is very high


To tackle the disadvatage of GridSearchCV ,sklearn library introduces new class called "RandomizedSearchCV"

RandomizedSearchCV
---

RandomizedSearchCV is a hyperparameter tuning technique in scikit-learn that randomly selects combinations of hyperparameters instead of trying every possible one (like GridSearchCV).

It is often faster and more efficient, especially when you have many hyperparameters or large search spaces.

üöÄ Why RandomizedSearchCV?
---
Because GridSearchCV becomes slow when:

    You have many parameters

    Each parameter has many possible values

    The model is computationally expensive

RandomizedSearchCV solves this by sampling a fixed number of random combinations.

üîç How RandomizedSearchCV Works
---
Step 1: You define parameter ranges, NOT all values

Example:

    param_dist = {
        'C': [0.1, 1, 10, 100, 1000],
        'gamma': [0.0001, 0.001, 0.01, 0.1],
        'kernel': ['rbf']
    }

Step 2: Instead of testing all combinations

    GridSearchCV tests all = 5 √ó 4 √ó 1 = 20 trials

    RandomizedSearchCV tests, say, 5 or 10 random combinations only.

Step 3: It applies cross-validation on each random combination

    ‚Üí Helps prevent overfitting.    

Step 4: It returns:

    Best hyperparameters

    Best model

    Best CV score

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False,n_iter=2)

rs.fit(iris_dataset.data,iris_dataset.target)
df=pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
df    
```

MODEL SELECTION USING HYPERPARAMETER TUNING
---

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model_params={

    'svm':{
    'model':SVC(gamma='auto'),
    'params':{    
        'C':[1,10,20],
        'kernel':['rbf','linear']
    }
},

    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{    
            'n_estimators':[1,5,10],
        }
},

    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{    
            'C':[1,5,10],
        }
}



}
```

```python
scores=[]
for model_name,model_parameter in model_params.items():
    clf=GridSearchCV(model_parameter['model'],model_parameter['params'],cv=5,return_train_score=False)
    clf.fit(iris_dataset.data,iris_dataset.target)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    }
    )
```

```python
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
