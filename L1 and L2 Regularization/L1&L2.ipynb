```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.shape
```

```python
df.drop([ 'Address','Date', 'Postcode', 'YearBuilt', 'Lattitude','Longtitude'],axis=1,inplace=True)
```

```python
# df['Bedroom2'].unique()
# df['Bathroom'].unique()
# df['Car'].unique()
# df['Landsize'].unique()
# df['BuildingArea'].unique
# df['Propertycount'].unique()
# df['Distance'].unique()
# df['CouncilArea'].unique()
# df['Regionname'].unique()
# df['Price'].unique()

```

```python
df['Bedroom2']=df['Bedroom2'].fillna(df['Bedroom2'].mean()).astype(int)
df['Landsize']=df['Landsize'].fillna(df['Landsize'].mean()).astype(int)
df['Bathroom']=df['Bathroom'].fillna(df['Bathroom'].mean()).astype(int)
df['Car']=df['Car'].fillna(df['Car'].mean()).astype(int)
df['BuildingArea']=df['BuildingArea'].fillna(df['BuildingArea'].mean()).astype(int)
df['Propertycount']=df['Propertycount'].fillna(df['Propertycount'].mean()).astype(int)
df['Distance']=df['Distance'].fillna(df['Distance'].mean()).round(2)
df['CouncilArea']=df['CouncilArea'].fillna(df['CouncilArea'].mode()[0])
df['Regionname']=df['Regionname'].fillna(df['Regionname'].mode()[0])
df['Price']=df['Price'].fillna(df['Price'].median()).round(2)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df=pd.get_dummies(df,drop_first=True,dtype=int)
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
lr_model.score(x_train,y_train)*100
```

```python
lr_model.score(x_test,y_test)*100
```

```python
from sklearn.linear_model import Lasso
lasso_model=Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_model.fit(x_train,y_train)
lasso_model.score(x_train,y_train)
```

```python
lasso_model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
ridge_model=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_model.fit(x_train,y_train)
ridge_model.score(x_train,y_train)
```

```python
ridge_model.score(x_test,y_test)
```
