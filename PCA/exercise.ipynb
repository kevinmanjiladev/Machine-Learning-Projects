```python
sns.boxplot(df['Oldpeak'])
df['Oldpeak_zscore']=(df['Oldpeak']-df['Oldpeak'].mean())/df['Oldpeak'].std()
df=df.loc[(df['Oldpeak_zscore']<3)&(df['Oldpeak_zscore']>-3)].copy()

df
```

```python
sns.boxplot(df['MaxHR'])
df['MaxHR_zscore']=(df['MaxHR']-df['MaxHR'].mean())/df['MaxHR'].std()
df=df.loc[(df['MaxHR_zscore']<3)&(df['MaxHR_zscore']>-3)].copy()

df
```

```python
sns.boxplot(df['FastingBS'])
df['FastingBS_zscore'] = (df['FastingBS'] - df['FastingBS'].mean()) / df['FastingBS'].std()
df=df.loc[(df['FastingBS_zscore']<3)&(df['FastingBS_zscore']>-3)].copy()

df
```

```python
sns.boxplot(df['Cholesterol'])
df['Cholesterol_zscore']=(df['Cholesterol']-df['Cholesterol'].mean())/df['Cholesterol'].std()
df=df.loc[(df['Cholesterol_zscore']<3)&(df['Cholesterol_zscore']>-3)].copy()

df
```

```python
sns.boxplot(df['RestingBP'])
df['RestingBP_zscore']=(df['RestingBP']-df['RestingBP'].mean())/df['RestingBP'].std()
df=df.loc[(df['RestingBP_zscore']<3)&(df['RestingBP_zscore']>-3)].copy()
df
```

```python
from sklearn.preprocessing import LabelEncoder
df = df.copy()
lab=LabelEncoder()
cols=['Sex','RestingECG','ExerciseAngina']
for i in cols:
    df[i]=lab.fit_transform(df[i])
```

```python
df = pd.get_dummies(df, columns=['ChestPainType','ST_Slope'], drop_first=True,dtype=int)
df
```

```python
df.dtypes
```

```python
df.drop(columns=['Oldpeak_zscore','MaxHR_zscore','FastingBS_zscore','Cholesterol_zscore','RestingBP_zscore'],axis=1,inplace=True)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
len(x)
```

```python
len(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
knn=KNeighborsClassifier()
lr=LogisticRegression()
rf=RandomForestClassifier()
svm=SVC()
```

```python
models=[knn,lr,rf,svm]
scores=[]
for model in models:
    model.fit(x_train,y_train)
    scores.append(model.score(x_test,y_test)*100)
print("Scores:", scores)
```

```python
x_train_pca,x_test_pca,y_train_pca,y_test_pca=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_pca=scaler.fit_transform(x_train_pca)
x_test_pca=scaler.transform(x_test_pca)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.97)
x_train_pca=pca.fit_transform(x_train_pca)
x_test_pca=pca.transform(x_test_pca)
```

```python
pca.n_components_
```

```python
pca.explained_variance_ratio_
```

```python
models=[knn,lr,rf,svm]
scores=[]
for model in models:
    model.fit(x_train_pca,y_train_pca)
    scores.append(model.score(x_test_pca,y_test_pca)*100)
print("Scores:", scores)
```
