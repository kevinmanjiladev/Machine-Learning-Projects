Here we are going to perform PCA

````python
from sklearn.decomposition import PCA
pca=PCA(0.97) # here we passed 95% because,We should get 95 % of useful information from the data and create new dimension by reducing unwanted features.
x_pca=pca.fit_transform(x)
x_pca
````

From above, it is clear that PCA reduces the dimension or features from 64 to 29 by neglecting unwanted features



# ðŸŒŸ **Where does PCA reduce features?**

PCA reduces features **AFTER** it transforms the data.

The reduction happens in this step:

```python
X_pca = pca.fit_transform(X)
```

This is the moment where the number of columns **shrinks**.

---

# ðŸ§  **Before PCA**

Your original data `X` has (example):

```
shape = (150, 4)
```

â†’ 150 rows
â†’ **4 features** (Sepal length, width, petal length, petal width)

---

# ðŸš€ **After PCA with 95% variance**

Suppose PCA decides that only **2 components** are needed to keep 95% info.

Then output becomes:

```
X_pca.shape = (150, 2)
```

### âœ”ï¸ 4 features â†’ 2 features

### âœ”ï¸ Reduction happens HERE

### âœ”ï¸ New features = PC1, PC2

---

# ðŸ”¥ **KEY POINT**

PCA does **not** reduce features during calculation.
It reduces features **after choosing how many PCs to keep**.

---

# ðŸ“Œ Visual Explanation (Super Simple)

### Original Features (4D):

```
f1   f2   f3   f4
```

### PCA creates new features (PCs):

```
PC1   PC2   PC3   PC4
```

But since you said:

```python
pca = PCA(0.95)
```

PCA checks variance and decides:

* PC1 â†’ gives 70% variance
* PC2 â†’ gives 22% variance
* PC3 â†’ gives 6% variance
* PC4 â†’ gives 2% variance

Total = 100%

To keep **95%**, PCA keeps:

âœ”ï¸ PC1
âœ”ï¸ PC2
âœ”ï¸ PC3

And **drops PC4** (only 2% importance).

---

# ðŸŸ¢ **Final Data After Feature Reduction**

Your final dataset becomes:

```
PC1   PC2   PC3     â† Only these 3 are kept
```

So:

### âœ”ï¸ You started with 4 features

### âœ”ï¸ You ended with 3 features

Feature reduction happened **here**.

---

# ðŸ’¡ **One-Line Summary**

ðŸ‘‰ PCA **first creates all new components**,

ðŸ‘‰ then **keeps only the top components** that explain enough variance (like 95%),

ðŸ‘‰ and **drops the rest**,

ðŸ‘‰ reducing the number of features.

---
