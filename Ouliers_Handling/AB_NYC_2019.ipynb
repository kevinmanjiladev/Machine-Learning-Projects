Step 1: Apply the 3-Sigma Rule
-------------------------------
<!-- most data (≈ 99.7%) lies within
mean ± 3 × std deviation -->
mean + 3 × std

= 112 + (3 × 58)

= 112 + 174

= 286

So according to statistics:

👉 Almost all normal values should be ≤ 286

Your max = 268, which is below 286.

So based on the 3-sigma rule:

❌ No extreme outlier beyond 3 std deviations.


Step 2: Compare with 75% (Interquartile method)
-----------------------------------------------
We compute IQR approach:

Q3=75%
Q1=25%

75% = 150,
25% = 65

Max = 268

IQR = Q3 - Q1

IQR = 150 - 65 = 85

Upper bound = Q3 + 1.5 × IQR 
            = 150 + 1.5 × 85 
            = 150 + 127.5 
            = 277.5
              =====

Upper bound = ~278
Max = 268 → lower than 278

So using IQR:

❌ No outlier beyond Q3 + 1.5 × IQR



Lower bound = Q1−1.5×IQR
-----------------------
Q1 = 65

Q3 = 150

IQR = Q3 - Q1

IQR = 150 - 65 = 85

Lower bound:
-----------
65−1.5(85)=65−127.5=−62.5


Lower bound = -62.5

But price per sqft cannot be negative in real estate.


```python
from scipy.stats import norm

plt.hist(df['price'],bins=20,rwidth=0.8,density=True)
plt.xlabel('Price')
plt.ylabel('Count')
rng=np.arange(df['price'].min(),df['price'].max(),0.01)
plt.plot(rng,norm.pdf(rng,df['price'].mean(),df['price'].std()))
```

```python
q1=df['price'].quantile(0.25)
q3=df['price'].quantile(0.75)
```

```python
IQR=q3-q1
IQR
```

```python
lower_limit=q1-(1.5*IQR)
upper_limit=q3+(1.5*IQR)
```

```python
lower_limit
```

```python
upper_limit
```

```python
outliers_df=df.loc[(df['price']<upper_limit)&(df['price']>lower_limit)]
```

```python
df.shape[0]-outliers_df.shape[0]
```

```python
plt.boxplot(outliers_df['price'])
```

```python
from scipy.stats import norm

plt.hist(outliers_df['price'],bins=20,rwidth=0.8,density=True)
plt.xlabel('Price')
plt.ylabel('Count')
rng=np.arange(outliers_df['price'].min(),outliers_df['price'].max(),0.01)
plt.plot(rng,norm.pdf(rng,outliers_df['price'].mean(),outliers_df['price'].std()))
```
