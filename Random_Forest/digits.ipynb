```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
df=pd.DataFrame(data=digits.data)
df
```

```python
df['target']=digits.target
df
```

```python
plt.gray()
plt.figure(figsize=(10,7))
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
x=df.drop(['target'],axis=1)
y=df['target']
```

```python
x
```

```python
y
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

✅ Why we use sklearn.ensemble

The module sklearn.ensemble in scikit-learn provides a collection of ensemble learning algorithms, such as:

RandomForestClassifier

RandomForestRegressor

GradientBoosting

AdaBoost

BaggingClassifier

VotingClassifier

StackingClassifier

These algorithms combine multiple models (like multiple decision trees) to improve accuracy and reduce overfitting.
