```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_wine
wine_dataset=load_wine()
```

```python
df=pd.DataFrame(data=wine_dataset.data,columns=wine_dataset.feature_names)
```

```python
df
```

```python
df['target']=wine_dataset.target
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['target'].unique()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)*100
score
```

```python
lab=[0,1,2]
cmd=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=lab)
cmd.plot()
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
y_pred1=model.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
cm1=confusion_matrix(y_test,y_pred1)
cm1
```

```python
score1=accuracy_score(y_test,y_pred1)*100
score1
```

```python
lab=[0,1,2]
cmd1=ConfusionMatrixDisplay(confusion_matrix=cm1,display_labels=lab)
cmd1.plot()
```
